# 2. Estadística para Data Science

La estadística es una herramienta fundamental en Data Science. Nos permite analizar, interpretar y visualizar datos, así como tomar decisiones informadas basadas en esos datos. En esta sección, exploraremos conceptos clave de estadística descriptiva, que nos ayudarán a resumir y comprender grandes conjuntos de datos.

## 2.1. Estadística Descriptiva

La estadística descriptiva se enfoca en describir y resumir los datos de manera informativa. Utiliza medidas numéricas y visualizaciones para proporcionar una imagen clara de los datos.

### 2.1.1. Medidas de tendencia central

Las medidas de tendencia central son valores que representan un conjunto de datos, proporcionando una idea de dónde se encuentra el centro de los datos.

#### 2.1.1.1. Media, mediana y moda

**Media:** La media es el promedio aritmético de un conjunto de valores. Se calcula sumando todos los valores y dividiendo por el número total de valores. Por ejemplo, para los datos [1, 2, 3, 4, 5], la media es (1+2+3+4+5)/5 = 3. La media es útil para conjuntos de datos simétricos, pero puede ser influenciada por valores atípicos.

**Mediana:** La mediana es el valor que divide el conjunto de datos en dos partes iguales. Si el número de valores es impar, es el valor del medio; si es par, es el promedio de los dos valores del medio. Por ejemplo, para los datos [1, 2, 3, 4, 5], la mediana es 3. La mediana es menos sensible a valores atípicos que la media.

**Moda:** La moda es el valor que aparece con mayor frecuencia en un conjunto de datos. Por ejemplo, para los datos [1, 2, 2, 3, 4], la moda es 2. En conjuntos de datos multimodales, puede haber más de una moda.

##### 2.1.1.1.1. Propiedades y comparación

- **Media:** Es útil para datos simétricos, pero afectada por valores atípicos.
- **Mediana:** Más robusta frente a valores atípicos, útil para datos sesgados.
- **Moda:** Útil para datos categóricos y para identificar valores frecuentes.

##### 2.1.1.1.2. Casos de uso en análisis de datos

- **Media:** Se usa en análisis financieros para calcular promedios de ingresos, gastos, etc.
- **Mediana:** En análisis de salarios para evitar el sesgo de valores atípicos.
- **Moda:** En análisis de mercado para identificar productos o servicios más populares.

#### 2.1.1.2. Implementación en Python

Python proporciona bibliotecas poderosas como NumPy y Pandas para el cálculo de estas medidas, así como Matplotlib y Seaborn para la visualización.

##### 2.1.1.2.1. Uso de NumPy y Pandas

```python
import numpy as np
import pandas as pd

# Datos de ejemplo
data = [1, 2, 2, 3, 4, 5, 5, 5, 6]

# Calcular media, mediana y moda usando NumPy
media = np.mean(data)
mediana = np.median(data)
moda = pd.Series(data).mode()[0]

print(f"Media: {media}, Mediana: {mediana}, Moda: {moda}")
```

##### 2.1.1.2.2. Visualización con Matplotlib y Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualización de los datos
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True)
plt.axvline(media, color='r', linestyle='--', label='Media')
plt.axvline(mediana, color='g', linestyle='-', label='Mediana')
plt.axvline(moda, color='b', linestyle='-', label='Moda')
plt.legend()
plt.title('Distribución de datos con Media, Mediana y Moda')
plt.show()
```

### 2.1.2. Medidas de dispersión

Las medidas de dispersión indican qué tan dispersos están los datos en un conjunto. Ayudan a comprender la variabilidad y el rango de los datos.

#### 2.1.2.1. Varianza, desviación estándar y rango intercuartílico

**Varianza:** La varianza mide la dispersión de los datos respecto a la media. Se calcula como la media de los cuadrados de las diferencias de cada valor con la media.

\[ \text{Varianza} = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2 \]

**Desviación estándar:** Es la raíz cuadrada de la varianza y proporciona una medida de dispersión en las mismas unidades que los datos originales.

\[ \text{Desviación estándar} = \sqrt{\text{Varianza}} \]

**Rango intercuartílico (IQR):** Es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1), y mide la dispersión de la mitad central de los datos.

\[ \text{IQR} = Q3 - Q1 \]

##### 2.1.2.1.1. Cálculo e interpretación

- **Varianza y desviación estándar:** Ayudan a entender la variabilidad de los datos. Una desviación estándar alta indica datos muy dispersos.
- **IQR:** Es útil para identificar y manejar valores atípicos.

##### 2.1.2.1.2. Coeficiente de variación

El coeficiente de variación (CV) es la relación entre la desviación estándar y la media, expresado como un porcentaje. Es útil para comparar la variabilidad entre diferentes conjuntos de datos.

\[ \text{CV} = \frac{\text{Desviación estándar}}{\text{Media}} \times 100\% \]

#### 2.1.2.2. Visualización con Python

La visualización ayuda a comprender mejor la dispersión de los datos y detectar valores atípicos.

##### 2.1.2.2.1. Boxplots y violinplots

```python
# Datos de ejemplo
data = np.random.normal(loc=0, scale=1, size=100)

# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data)
plt.title('Boxplot de datos')
plt.show()

# Violinplot
plt.figure(figsize=(10, 6))
sns.violinplot(data=data)
plt.title('Violinplot de datos')
plt.show()
```

##### 2.1.2.2.2. Histogramas y densidad kernel

```python
# Histograma
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=10, kde=True)
plt.title('Histograma de datos con KDE')
plt.show()
```

En resumen, las medidas de tendencia central y dispersión son herramientas fundamentales para el análisis de datos en Data Science. Estas medidas, junto con visualizaciones efectivas, permiten una comprensión profunda y clara de los datos, lo que facilita la toma de decisiones informadas.
